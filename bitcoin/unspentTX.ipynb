```python
import json
import requests
```

```python
def utxo(address):
    resp = requests.get('https://blockchain.info/unspent?active=%s' % address)
    utxo_set = json.loads(resp.text)["unspent_outputs"]
    for utxo in utxo_set:
        print(f"hash: {utxo['tx_hash']}\nUnspent outputs: {utxo['tx_output_n']}\nValue: {utxo['value']}")
```

```python
utxo('12tkqA9xSoowkzoERHMWNKsTey55YEBqkv')
```
